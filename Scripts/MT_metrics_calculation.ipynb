# **CamelBERT similarity**

```python
import torch
from transformers import AutoModel, AutoTokenizer
from sklearn.metrics.pairwise import cosine_similarity

def get_sentence_embedding(sentence, model, tokenizer):
    # Tokenize the input sentence
    inputs = tokenizer(sentence, return_tensors='pt', padding=True, truncation=True)

    # Get the hidden states from the model
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the embeddings from the last hidden state
    # We take the mean of all token embeddings to get the sentence embedding
    embeddings = outputs.last_hidden_state.mean(dim=1)

    return embeddings

def calculate_cosine_similarity(embedding1, embedding2):
    # Calculate cosine similarity
    similarity = cosine_similarity(embedding1.numpy(), embedding2.numpy())
    return similarity[0][0]
```

```python
# Load pre-trained model and tokenizer
model_name = "CAMeL-Lab/bert-base-arabic-camelbert-msa"  # You can use any pre-trained transformer model here
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
import pandas as pd

df = pd.read_csv('/content/ALG-MSA-Diac-POS.csv', sep=',')
df.head()
```

With POS insertion

```python
similarity = 0
df.fillna('/', inplace=True)
for i in range(len(df)):
  # Calculate cosine similarity
  similarity += calculate_cosine_similarity(get_sentence_embedding(df['ALG-MSA-DIACT-gemma2:9b-instruct-q5_0'][i], model, tokenizer), get_sentence_embedding(df['Padic-MSA'][i], model, tokenizer))

print(f"Cosine Similarity: {similarity/len(df)}")
```
