```python
import json
import ntpath
import glob
import os
import numpy as np
from openai import OpenAI
import pandas as pd
import re

df = pd.read_csv('/content/ALG PADIC Diac POS.csv', sep=',')
df.head()
```

```python
from tqdm import tqdm
client = OpenAI(
    # This is the default and can be omitted
    api_key='xxx',
)

translated = []

for i in tqdm(range(len(df))):
  prompt=     """
          Translate a given sentence from Algerian Dialect to Modern Standard Arabic.

                  Your input will include a sentence in Algerian Dialect along with the diacritization for each word and also the part of speech of each word.
                  Use this information to accurately translate the sentence into Modern Standard Arabic .
                  The translation should convey the original meaning while adhering to Modern Standard Arabic norms.

                  # Steps

                  1. **Understand Context**: Analyze the provided sentences, diacritization, and part of speech to grasp the overall context and structure.
                  2. **Word-by-Word Translation**: Consider each word and its diacritization and grammatical class to find an appropriate Modern Standard Arabic equivalent.
                  3. **Reconstruction**: Adjust the translated words to form a grammatically correct and coherent sentence in Modern Standard Arabic.
                  4. **Review for Meaning**: Ensure that the translated sentence accurately reflects the meaning and nuances of the original Algerian Dialect sentence.

                  # Output Format

                  - Provide a single translated sentence in Modern Standard Arabic.


                  *(Note: The example is shorter than typical to illustrate formatting. In practice, take into account longer sentences as well.)*
                  OUTPUT: JUST THE TRANSLATED SENTENCE IN Algerian Dialect  WITHOUT ANY ADDITIONAL TEXT.

                  """ + 'Sentence with diacritics' + df['ALG-DIACT-Sentences'][i] + '\n' + 'Sentece with part of speech' + df['Padic-ALG-POS'][i]

  response = client.chat.completions.create(
              model="gpt-4o-mini-2024-07-18",
              #response_format={"type":"json_object"},
              messages=[
                    {"role": "user", "content": prompt},
  ])

  translated.append(response.choices[0].message.content)
df['ALG-MSA-DIACT-POS-gpt-4o-mini-2024-07-18'] = translated
#gpt-4-1106-preview
#gpt-4o-mini-2024-07-18
#chatgpt-4o-latest
```

```python
import os
import getpass
from langchain_anthropic import ChatAnthropic
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
import json
import tqdm
from tqdm import tqdm

from langchain_google_genai import ChatGoogleGenerativeAI

# If there is no env variable set for API key, you can pass the API key
# to the parameter `google_api_key` of the `ChatGoogleGenerativeAI` function:
# `google_api_key="key"`.

from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    google_api_key='xxx'
    # other params...
)
#gemini-1.5-flash
#gemini-1.5-flash-8b
#gemini-1.5-pro
```

```python
translated = []
system_template =    """
 Translate a given sentence from Algerian Dialect to Modern Standard Arabic.

                  Your input will include a sentence in Algerian Dialect along with the diacritization for each word and also the part of speech of each word.
                  Use this information to accurately translate the sentence into Modern Standard Arabic .
                  The translation should convey the original meaning while adhering to Modern Standard Arabic norms.

                  # Steps

                  1. **Understand Context**: Analyze the provided sentences, diacritization, and part of speech to grasp the overall context and structure.
                  2. **Word-by-Word Translation**: Consider each word and its diacritization and grammatical class to find an appropriate Modern Standard Arabic equivalent.
                  3. **Reconstruction**: Adjust the translated words to form a grammatically correct and coherent sentence in Modern Standard Arabic.
                  4. **Review for Meaning**: Ensure that the translated sentence accurately reflects the meaning and nuances of the original Algerian Dialect sentence.

                  # Output Format

                  - Provide a single translated sentence in Modern Standard Arabic.


                  *(Note: The example is shorter than typical to illustrate formatting. In practice, take into account longer sentences as well.)*
                  OUTPUT: JUST THE TRANSLATED SENTENCE IN Algerian Dialect  WITHOUT ANY ADDITIONAL TEXT.
                  """

for i in tqdm(range(len(df))):
  messages = [
    (
        "system",
        system_template,
    ),
    ("human", '\n Sentence with diacritics: \n' + df['ALG-DIACT-Sentences'][i] + '\n' + 'Sentece with part of speech: \n' + df['Padic-ALG-POS'][i]),
  ]
  ai_msg = llm.invoke(messages)
  translated.append(ai_msg.content)


df['ALG-MSA-DIACT-POS-gemini-1.5-flash'] = translated
df.to_csv('ALG-MSA-Diac-POS.csv', index=False)
print('completed successfully')
```
